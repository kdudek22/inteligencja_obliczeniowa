```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_1 = pd.read_csv("./data/1.csv")
df_2 = pd.read_csv("./data/2.csv")
df_3 = pd.read_csv("./data/3.csv")
df_4 = pd.read_csv("./data/4.csv")
```

```python
plt.plot(df_1["Step"], df_1["Value"], color="blue", label="1")
plt.plot(df_2["Step"], df_2["Value"], color="red", label="2")
plt.plot(df_3["Step"], df_3["Value"], color="purple", label="3")
plt.plot(df_4["Step"], df_4["Value"], color="green", label="4")
plt.legend()
plt.savefig("compare.png")
```
